## 3. Cleaning DataSet

### 3.1 Removing Outliers

```python
# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['number_of_applications'].quantile(0.25)
Q3 = df['number_of_applications'].quantile(0.92)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Identify outliers using the 1.5 * IQR rule
outliers = df[(df['number_of_applications'] < (Q1 - 1.5 * IQR)) | (df['number_of_applications'] > (Q3 + 1.5 * IQR))]

# Filter the DataFrame to exclude outliers
df = df[~((df['number_of_applications'] < (Q1 - 1.5 * IQR)) | (df['number_of_applications'] > (Q3 + 1.5 * IQR)))]
```

### 3.2 Checking the missing value

```python
df.isnull().sum()
```

```python
# Create a heatmap to visualize missing values
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# List of values to drop from the specified column
value_to_drop = ['Active', 'MIA', 'Deferred', 'Break']

# Name of the column to filter
column_name = 'pathrise_status'

# Drop the rows with specified values from the DataFrame
df = df[~df[column_name].isin(value_to_drop)]

# Reset the index if needed
df = df.reset_index(drop=True)
```

### 3.3 Dealing with missing values

```python
# Fill missing values in 'cohort_tag' column with 'JAN19A'
df['cohort_tag'].fillna('JAN19A', inplace=True)

# Mapping to convert cohort month abbreviations to numeric values
cohort_to_month_mapping = {
    'JAN': '01', 'FEB': '02', 'MAR': '03', 'APR': '04', 'MAY': '05', 'JUN': '06',
    'JUL': '07', 'AUG': '08', 'SEP': '09', 'OCT': '10', 'NOV': '11', 'DEC': '12'
}

# Replace cohort month abbreviations with numeric values using the mapping dictionary
df['cohort_tag'] = df['cohort_tag'].replace(cohort_to_month_mapping, regex=True)

# Mapping to convert cohort day codes to numeric values
cohort_to_day_mapping = {
    'A': '01', 'B': '15', 'C': '25', 'a': '01'
}

# Replace cohort day codes with numeric values using the mapping dictionary
df['cohort_tag'] = df['cohort_tag'].replace(cohort_to_day_mapping, regex=True)

# Convert 'cohort_tag' column to datetime format
df['cohort_tag'] = pd.to_datetime(df['cohort_tag'], format='%m%y%d')

# Extract year and month from the 'cohort_tag' column
year = df['cohort_tag'].dt.year
month = df['cohort_tag'].dt.month

# Insert 'year' and 'month' columns at specific positions in the DataFrame
df.insert(2, 'year', year)
df.insert(3, 'month', month)
```

```python
# Drop the specified columns from the DataFrame
df.drop(['pathrise_status', 'id', 'cohort_tag'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

### 3.4 Filling the missing values

```python
# Fill missing values in 'employment_status' column with 'unknown'
df['employment_status'].fillna('unknown', inplace=True)

# Fill missing values in 'highest_level_of_education' column with "Bachelor's Degree"
df['highest_level_of_education'].fillna("Bachelor's Degree", inplace=True)

# Fill missing values in 'length_of_job_search' column with 'Less than one month'
df['length_of_job_search'].fillna('Less than one month', inplace=True)

# Fill missing values in 'biggest_challenge_in_search' column with 'Hearing back on my applications'
df['biggest_challenge_in_search'].fillna('Hearing back on my applications', inplace=True)

# Fill missing values in 'professional_experience' column with 'unknown'
df['professional_experience'].fillna('unknown', inplace=True)

# Fill missing values in 'work_authorization_status' column with 'Citizen'
df['work_authorization_status'].fillna('Citizen', inplace=True)

# Fill missing values in 'gender' column with 'Decline to Self Identify'
df['gender'].fillna('Decline to Self Identify', inplace=True)

# Fill missing values in 'race' column with 'Decline to Self Identify'
df['race'].fillna('Decline to Self Identify', inplace=True)
```

```python
# Calculate the mean of the 'number_of_interviews' column
df['number_of_interviews'].mean()

# Fill missing values in 'number_of_interviews' column with the mean value
df['number_of_interviews'].fillna(2, inplace=True)
```

```python
# Calculate the mean program duration for each employment status group
mean_by_employment_status = df.groupby('employment_status')['program_duration_days'].transform('mean').round()

# Fill missing values in 'program_duration_days' column with the corresponding mean value
filled_program_duration = df['program_duration_days'].fillna(mean_by_employment_status)

# Update the 'program_duration_days' column with the filled values
df['program_duration_days'] = filled_program_duration
```

## 5. Run the models

### 5.1 Import necessery libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from sklearn.metrics import classification_report
import warnings
from sklearn.exceptions import DataConversionWarning
```

### 5.2 Divide and Scale Dataset

```python
# Divide our Dataset to X and Y
X = df.drop('placed', axis=1)
y = df['placed']
```

```python
# Create a StandardScaler instance
sc = StandardScaler()

# Fit and transform the features using the StandardScaler
x_scaled = sc.fit_transform(X)

# x_scaled now contains the standardized features
x_scaled
```

```python
# Get the original column names
col = X.columns

# Create a new DataFrame 'X' with the standardized features and original column names
X = pd.DataFrame(x_scaled, columns=col)

# 'X' now contains the standardized features with original column names
X
```

```python
# Split data and set the test size
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

### 5.3 Define parameters and fit the models

```python
# Define classifiers
knn = KNeighborsClassifier()
svm = SVC()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
lr = LogisticRegression()

# Define parameter grids for GridSearchCV
param_grid_knn = {'n_neighbors': [3, 5, 7, 9]}
param_grid_svm = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
param_grid_dt = {'criterion': ['gini', 'entropy'], 'max_depth': [3, 5, 7, 9]}
param_grid_rf = {'n_estimators': [50, 100, 150, 200], 'max_depth': [3, 5, 7, 9]}
param_grid_lr = {'C': [0.1, 1, 10]}

# Perform GridSearchCV for each classifier
grid_knn = GridSearchCV(knn, param_grid_knn, cv=10)
grid_svm = GridSearchCV(svm, param_grid_svm, cv=10)
grid_dt = GridSearchCV(dt, param_grid_dt, cv=10)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=10)
grid_lr = GridSearchCV(lr, param_grid_lr, cv=10)

# Fit the models
grid_knn.fit(X_train, y_train)
grid_svm.fit(X_train, y_train)
grid_dt.fit(X_train, y_train)
grid_rf.fit(X_train, y_train)
grid_lr.fit(X_train, y_train)
```

```python
# Create instances of classifiers with best hyperparameters
knn = KNeighborsClassifier(n_neighbors=grid_knn.best_params_['n_neighbors'])

svm = SVC(C=grid_svm.best_params_['C'], kernel=grid_svm.best_params_['kernel'])

dt = DecisionTreeClassifier(criterion=grid_dt.best_params_['criterion'], max_depth=grid_dt.best_params_['max_depth'])

rf = RandomForestClassifier(n_estimators=grid_rf.best_params_['n_estimators'], max_depth=grid_rf.best_params_['max_depth'])

lr = LogisticRegression(C=grid_lr.best_params_['C'])
```

```python
# Fit each model to the training data
knn.fit(X_train, y_train)

svm.fit(X_train, y_train)

dt.fit(X_train, y_train)

rf.fit(X_train, y_train)

lr.fit(X_train, y_train)
```

### 5.4 Run and categorize the models

```python
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DataConversionWarning)

## Initializing models, model_names and evaluation metrics
models = [knn, svm, dt, rf, lr]
model_names = ['KNN', 'SVM', 'DT', 'RF', 'LR']
accuracies = []
confusion_matrices = []
fprs = []
tprs = []
aucs = []

## Loop through models to make predictions using each model
for model in models:
    y_pred = model.predict(X_test)
    ## store evaluation metrics of each model
    accuracies.append(accuracy_score(y_test, y_pred))
    confusion_matrices.append(confusion_matrix(y_test, y_pred))
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    fprs.append(fpr)
    tprs.append(tpr)
    aucs.append(auc(fpr, tpr))
```

```python
# Loop through models
for i in range(len(models)):
    print(f"Model: {model_names[i]}")
    
    # Print accuracy score
    print(f"Accuracy Score: {accuracies[i]:.2f}")
    
    # Display confusion matrix heatmap
    sns.heatmap(confusion_matrices[i], annot=True, cmap="Oranges", fmt='d')
    plt.title(f'Confusion Matrix - {model_names[i]}')
    plt.show()
    
    # Make predictions using the model
    y_pred = models[i].predict(X_test)
    
    # Print classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred), "\n")
```

### 5.5 ROC

```python
# Create a figure
plt.figure(figsize=(8, 6))

# Plot ROC curves for each model
for i in range(len(models)):
    plt.plot(fprs[i], tprs[i], label=model_names[i] + " (AUC = {:.2f})".format(aucs[i]))

# Plot the diagonal line (random guessing)
plt.plot([0, 1], [0, 1], 'k--', color="red")

# Set plot properties
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")

# Display the plot
plt.show()
```
